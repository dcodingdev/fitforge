```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from scipy.sparse import csr_matrix
```

```python
workouts = pd.read_csv("../data/megaGymDataset.csv")
ratings = pd.read_csv("../data/user_workout_ratings_full_bias.csv")
```

```python
le_type = LabelEncoder()
le_body = LabelEncoder()
le_equip = LabelEncoder()
le_level = LabelEncoder()
```

```python
workouts["Type_enc"] = le_type.fit_transform(workouts["Type"])
workouts["BodyPart_enc"] = le_body.fit_transform(workouts["BodyPart"])
workouts["Equipment_enc"] = le_equip.fit_transform(workouts["Equipment"])
workouts["Level_enc"] = le_level.fit_transform(workouts["Level"])
```

```python
pivot = ratings.pivot_table(index="workout_id", columns="user_id", values="rating").fillna(0)
sparse_matrix = csr_matrix(pivot.values)
```

```python
print("Pivot shape:", pivot.shape)
print("Feature engineering complete. Matrix ready for model training.")
```

```python
pivot.to_pickle("../models/pivot.pkl")
import joblib
joblib.dump(sparse_matrix, "../models/sparse_matrix.pkl")
```
